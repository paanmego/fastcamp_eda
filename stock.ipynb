```python
import pandas as pd
```

```python
df = pd.read_csv('./nasdaq_data.csv')
fund = pd.read_csv('./DFF.csv')

fund
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

start_date = '1995-01-20'
end_date = '2024-04-28'

day_filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
fund_filtered_df = fund[(fund['DATE'] >= start_date) & (fund['DATE'] <= end_date)]

#fig = go.Figure()
fig = make_subplots(rows=3, cols=1, shared_xaxes=True, vertical_spacing=0.02, row_heights=[0.5, 0.2, 0.3], subplot_titles=('Nasdaq Index', 'Volume', 'Fund'))
fig.add_trace(go.Candlestick(x=day_filtered_df['Date'], open=day_filtered_df['Open'], high=day_filtered_df['High'], low=day_filtered_df['Low'], close=day_filtered_df['Close'], name='Nasdaq Index'), row=1, col=1)
fig.add_trace(go.Bar(x=day_filtered_df['Date'], y=day_filtered_df['Volume'], marker_color='blue', name='Stock Volume'), row=2, col=1)
fig.add_trace(go.Line(x=fund_filtered_df['DATE'], y=fund_filtered_df['DFF'], marker_color='gray', name='fund'), row=3, col=1)


fig.update_layout(height=1200, width=1000, title_text="Stock Price and Volume")

fig.show()
```

```python
fund_filtered_df.rename(columns={'DATE': 'Date'}, inplace=True)

merged_df = pd.merge(day_filtered_df, fund_filtered_df, on='Date', how='inner')
merged_df
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
import yfinance as yf


def get_sp500_tickers():
    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')
    table = soup.find('table', {'class': 'wikitable sortable'})
    
    # 테이블에서 심볼 추출
    tickers = []
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text.strip()
        tickers.append(ticker)
    
    return tickers

#Symbol list
sp500_tickers = get_sp500_tickers()
print(sp500_tickers)

start_date = "2022-01-01"
end_date = "2022-12-31"

# up_tickers = []
# for ticker in sp500_tickers:
#     data = yf.download(ticker, start=start_date, end=end_date)
#     if not data.empty:
#         # 시작일과 종료일의 'Close' 가격 비교
#         start_price = data.iloc[0]['Close']
#         end_price = data.iloc[-1]['Close']
        
#         if end_price > start_price:
#             up_tickers.append(ticker)

#print("Rising stocks :", up_tickers)
```
