```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random

import datetime as dt
```

```python
import pandas as pd
```

```python
import json

with open('./top_10_code.json', 'rt', encoding='UTF8') as file:
    data = json.load(file)
#5월 2일 기준 거래량, 시가총액 상위 10개 kosdaq 종목
```

```python
browser = webdriver.Chrome()
```

```python
url = 'https://finance.naver.com/'
```

```python
browser.get(url)
```

```python
for asdf in range(0, 1):
    browser.find_element(By.CLASS_NAME, 'search_input').click()
    browser.find_element(By.CLASS_NAME, 'search_input').send_keys(data['top_10_trading_volume'][asdf])
    browser.find_element(By.CLASS_NAME, 'search_input').send_keys(Keys.ENTER)
    time.sleep(1)

    tmp_url = browser.find_element(By.CLASS_NAME, 'section.new_bbs')\
        .find_element(By.CLASS_NAME, 'sub_section.right')\
        .find_element(By.CLASS_NAME, 'more').get_attribute('href')
    
    page_no = 100
    i = 13
    tmp_url2 = tmp_url + '&page=' + str(page_no)
    browser.get(tmp_url2)
    time.sleep(1)

    data_list = []

    while True:

        table = browser.find_element(By.CLASS_NAME, 'type2') # 페이지의 테이블 추출
        tbody = table.find_element(By.TAG_NAME, "tbody") # 테이블의 내용만 추출
        rows = tbody.find_elements(By.TAG_NAME, "tr") # 내용을 행단위로 끊어 추출

        for row in rows[2:len(rows)-1]: # 각 행에 대해 작업 진행
            
            this_row = row.find_elements(By.TAG_NAME, 'td') # 각 행의 요소들을 추출 
            
            if len(row.text) == 0: # 행의 내용이 없을시 무시
                continue

            if dt.datetime.strptime(this_row[0].text, "%Y.%m.%d %H:%M") < dt.datetime(2020, 1, 1): # 게시물의 작성일 이 end_date 보다 낮다면 종료
                break

            

            if this_row[0].text != '' and dt.datetime.strptime(this_row[0].text, "%Y.%m.%d %H:%M") < dt.datetime(2022, 9, 1) and\
                dt.datetime.strptime(this_row[0].text, "%Y.%m.%d %H:%M") >= dt.datetime(2020, 1, 1): # 해당 글의 작성일이 코로나 시기 사이라면 데이터 추출
                
                created_date = this_row[0].text
                title = this_row[1].text
                writer = this_row[2].text
                views = this_row[3].text
                like = this_row[4].text
                dislike = this_row[5].text

                data_list.append({
                    'create_date' : created_date,
                    'title' : title,
                    'writer' : writer,
                    'views' : views,
                    'like' : like,
                    'dislike' : dislike
                })

        if dt.datetime.strptime(this_row[0].text, "%Y.%m.%d %H:%M") < dt.datetime(2020, 1, 1): # 게시물의 작성일 이 end_date 보다 낮다면 종료
                break

        # if page_no < 100:
        #     # 페이지를 이동
        #     page_no = page_no + 1
        #     tmp_url3 = tmp_url + '&page=' + str(page_no)
        #     browser.get(tmp_url3)
        #     time.sleep(1)
        # else:
        # 페이지 이동 클릭
        xpath = f'//*[@id="content"]/div[2]/table[2]/tbody/tr/td[2]/table/tbody/tr/td[{i}]/a'
        if i == 13:
            i = 4
        else:
            i = i + 1
        browser.find_element(By.XPATH, xpath).click()
        time.sleep(1)


    new_df = pd.DataFrame(data_list)

    new_df.to_csv('./covid_19_volume/code_'+ data['top_10_trading_volume'][asdf] +'.csv', sep=',', index=False)
```
